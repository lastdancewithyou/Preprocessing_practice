# 데이터 전처리
- tidy data 형식으로 만들기

깔끔한 데이터(tidy data)는 데이터셋의 의미를 구조에 매칭하는 표준적인 방식이다. 행, 열, 테이블이 관측점, 변수, 형식에 매칭되는 방식에 따라 데이터셋이 깔끔하거나 깔끔하지 않은 것으로 나뉜다. 깔끔한 데이터(tidy data) 는 결국 데이터 분석을 쉽게 할 수 있는 데이터로 볼 수 있다.

```python
df_raw.columns
```

```python
# id_vars 여러 컬럼 지정 시 리스트 형태로 지정해줘야 함.
df = df_raw.melt(id_vars= ["국가(대륙)별", "상품군별", "판매유형별"], var_name = "기간", value_name = "백만원")
df.shape
```

```python
df.head()
```

```python
df.info()
```

## 기간에서 연도 분리하기

```python
int("2014.1/4".split(".")[0])
```

```python
df["연도"] = df["기간"].map(lambda x : int(x.split(".")[0]))
df.head()
```

```python
# 분기 분리하기
("2014.1/4".split(".")[1]).split("/")[0]
```

```python
# 분기 컬럼 만들기
df["분기"] = df["기간"].map(lambda x : int(x.split(".")[1].split("/")[0]))
df.head()
```

```python
# - 문자를 결측치 처리하고 float 타입으로 변경하기
df["백만원"] = df["백만원"].replace("-", pd.np.nan).astype(float)
df.head()
```

## 필요 없는 데이터 제거하기

```python
df
```

```python
df = df[(df["국가(대륙)별"] != "합계") & (df["상품군별"] != "합계")].copy()
df.info()
```

```python
df.isnull().sum()
```
